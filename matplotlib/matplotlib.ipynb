```python
from matplotlib import pyplot as plt
from matplotlib import style
%matplotlib inline
```

```python
plt.plot([1,2,3,4],[2,5,7,9])        # ploting thr list on graph
plt.show()
```

```python
plt.plot([1,2,3,4],[2,5,7,9])
plt.title("Sample Data")         # adding title to the graph plot
plt.xlabel("X label")                # adding x label to the graph plot
plt.ylabel("Y label")                    # adding y label to the graph plot
plt.show()
```

```python
a = [1,5,7]
b = [3,4,9]
c = [5,4,6]
d = [5,6,5]
```

```python
# simple ploting 

plt.plot(a,b)
plt.plot(c,d)
plt.show()
```

```python
# used ggplot style to plot

style.use('ggplot')
plt.plot(a,b)
plt.plot(c,d)
plt.show()
```

```python
# used legend functio to show the legend to plot

style.use('ggplot')
plt.plot(a,b, label = 'First')
plt.plot(c,d, label = 'Second')
plt.legend()
plt.show()
```

```python
# used linewidth to increase the size of lines in plot

style.use('ggplot')
plt.plot(a,b, label = 'First', linewidth = 4)
plt.plot(c,d, label = 'Second', linewidth = 3)
plt.legend()
plt.show()
```

```python
# scatter plot using plt.scatter

style.use('ggplot')
plt.scatter(a,b, label = 'First', linewidth = 4)
plt.scatter(c,d, label = 'Second', linewidth = 3)
plt.legend()
plt.show()
```

```python
# pie chart

plt.pie(a)
plt.show()
```

```python
# bar chart

plt.bar(a,b)
plt.show
```

```python
# pie chart in details

slices = [7,2,5,11]
activities = ['sleeping', 'eating', 'meeting','playing']
plt.pie(slices, labels = activities, startangle = 90, shadow = True, explode =(0.2,0.1,0.2,0.01))
plt.show()
```
